```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.standard_normal(2)
x
```

```python
y = np.random.standard_normal(2)
y
```

```python
# Distance between the two points:
np.linalg.norm(x-y)
```

```python
d = np.arange(2, 20)
distance = []
for dim in d:
    x = np.random.standard_normal(dim)
    y = np.random.standard_normal(dim)
    distance.append(np.linalg.norm(x-y))
    
plt.scatter(d, distance)
plt.xlabel("dimension of space")
plt.ylabel("distance")
```
